# Proyecto Telecom X para Alura LATAM

## Importación de bibliotecas, datos y creación de dataframe.

```python
import requests
import pandas as pd

url = "https://raw.githubusercontent.com/sthemonica/alura-voz/refs/heads/main/Dados/Telco-Customer-Churn.json"

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)

df
```

```python
#Normalizamos, es decir, se hace un df con columnas donde antes había diccionarios
df_normalizado = pd.json_normalize(data)
df_normalizado
```

```python
df_normalizado.info()
```

```python
#Se guarda el DF normalizado para reusarlo en otro momento.
df_normalizado.to_csv('df_normalizado.csv', index=False)
```

```python
# Verifique si hay valores únicos en cada columna
for col in df_normalizado.columns:
  print(f"valores únicos en la columna'{col}': {df_normalizado[col].nunique()}")
  if df_normalizado[col].nunique() < 50: # Imprima los valores solamente si no hay muchos
    print(df_normalizado[col].unique())
    print('-' * 50)  # Línea divisoria para mejor visualización
```

```python
#Duplicados
print("Numero de duplicados", df_normalizado.duplicated().sum())
```

```python
# Nulos
print("Numero de Nulos", df_normalizado.isnull().sum())
```

```python
df_normalizado.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
df_normalizado['account.Charges.Total'] = pd.to_numeric(df_normalizado['account.Charges.Total'], errors='coerce')
df_normalizado['account.Charges.Total'].dtype
```

```python
# Se eliminan filas vacías, deben de ser: 224

df_normalizado = df_normalizado[df_normalizado['Churn'].str.strip() != '']
print("Número de filas después de eliminar las vacías en 'Churn':", len(df_normalizado))
```

```python
df_normalizado.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
df_normalizado.describe()
```

```python
df_normalizado['account.Charges.Daily'] = df_normalizado['account.Charges.Monthly'] / 30
df_normalizado['account.Charges.Daily']
```

```python
df_normalizado = df_normalizado.drop('account.Charges.Daily', errors='ignore')
```

```python
df_normalizado.describe()
```

```python
#Df normalizado ya con la nueva columna de Cargos diarios
df_normalizado.to_csv('df_normalizado_charges_daily.csv', index=False)
```

```python
# Diccionario de renombramiento: inglés -> español
columnas_renombradas = {
    'customerID': 'id_cliente',
    'Churn': 'evadido',
    'customer.gender': 'genero',
    'customer.SeniorCitizen': 'es_adulto_mayor',
    'customer.Partner': 'tiene_pareja',
    'customer.Dependents': 'tiene_dependientes',
    'customer.tenure': 'meses_en_empresa',

    'phone.PhoneService': 'servicio_telefonico',
    'phone.MultipleLines': 'lineas_multiples',

    'internet.InternetService': 'tipo_internet',
    'internet.OnlineBackup': 'respaldo_en_linea',
    'internet.DeviceProtection': 'proteccion_dispositivo',
    'internet.TechSupport': 'soporte_tecnico',
    'internet.StreamingTV': 'tv_streaming',
    'internet.StreamingMovies': 'peliculas_streaming',

    'account.Contract': 'tipo_contrato',
    'account.PaperlessBilling': 'facturacion_electronica',
    'account.PaymentMethod': 'metodo_pago',
    'account.Charges.Monthly': 'cargo_mensual',
    'account.Charges.Total': 'cargo_total'
}

# Renombrar columnas
df_normalizado.rename(columns=columnas_renombradas, inplace=True)

# Confirmar renombramiento
df_normalizado.columns
```

```python
df_normalizado.to_csv('df_normalizado_renombrado.csv', index=False)
```

```python
# Detectar automáticamente columnas binarias tipo 'Yes/No'
#Si existe un tercer valor, no transforma los datos
cols_binarias_auto = [
    col for col in df_normalizado.columns
    if df_normalizado[col].dropna().astype(str).str.lower().isin(['yes', 'no']).all()
]

# Convertir 'yes' a 1 y 'no' a 0
for col in cols_binarias_auto:
    df_normalizado[col] = df_normalizado[col].str.lower().map({'yes': 1, 'no': 0})

df_normalizado
```

```python
#Se genera nuevo df en csv
df_normalizado.to_csv('df_normalizado_binarias.csv', index=False)
```
